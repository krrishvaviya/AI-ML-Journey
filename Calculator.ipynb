```python
num1 = float(input ("enter first number"))
num2 = float(input("input second number"))
op = input("enter operation to be done")

if op == "+":
  print(num1 + num2)
elif op =="-":
  print(num1 - num2)
elif op == "/":
  print(num1 / num2)
elif op == "*":
  print(num1 * num2)
```
