* CREATING A SIMPLE CALCULATOR*

```python
num1  = input ("enter 1 st number ")
num2 = input (" input second number ")
result = (int(num1) + int(num2))               # here it takes input as string if we dont add "int" before both variable it will treate the input as string and answer 45
print (f"the addition is = {result} ")
```

```python
num1  = input ("enter 1 st number ")
num2 = input (" input second number ")
result = (float(num1) + float(num2))               # here it takes input as string if we dont add "float" before both variable it will treate the input as string and answer 45
print (f"the addition is = {result} ")             # int inly numbers no decimal float both decimal and normal numbers
```
