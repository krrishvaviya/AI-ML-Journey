if else statement each ef ,elif has a condition id the condition satisfies it prints or works as per that and else works as default

```python
is_male = False
is_tall = False

if is_male and is_tall:
  print("you are tall and a male")
elif is_male and not(is_tall):
  print("male but not tall")
elif not(is_male) and is_tall:
  print("not male but tall")
else:
  print("you are neither male nor tall")
```
